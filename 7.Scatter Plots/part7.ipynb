## Scatter Plots

```python
# Import necessary libraries
import pandas as pd
from matplotlib import pyplot as plt

# Load dataset from CSV file
data = pd.read_csv('2019-05-31-data.csv')

# Apply a dark theme style for better contrast
plt.style.use("dark_background")

# Extract relevant columns from dataset
view_count = data['view_count']   # Total views of the video
likes = data['likes']             # Total likes of the video
ratio = data['ratio']             # Like-to-dislike ratio

# Create scatter plot
# - x-axis: view_count
# - y-axis: likes
# - c=ratio  color of points based on ratio values
# - cmap='rainbow'  rainbow color gradient
# - s=50  marker size
# - edgecolor='black'  outline around markers
# - linewidth=1  thickness of outline
# - alpha=0.75  transparency level
plt.scatter(view_count, likes, c=ratio, cmap='rainbow', s=50,
            edgecolor='black', linewidth=1, alpha=0.75)

# Add color bar to explain the color mapping
cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

# Set log scale for both axes
# (useful when data has very large range of values)
plt.xscale('log')
plt.yscale('log')

# Add chart title and axis labels
plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

# Adjust spacing to prevent labels from overlapping
plt.tight_layout()

# Save the plot as an image file
plt.savefig("scatterplot.png")

# Display the plot on screen
plt.show()
```
