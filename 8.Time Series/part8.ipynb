## Plotting Plotting Time Series Data

```python
import pandas as pd
import matplotlib.pyplot as plt


plt.style.use("dark_background") 

# Load dataset from CSV file
data = pd.read_csv('data.csv')

# Convert the "Date" column from string format to datetime objects
data["Date"] = pd.to_datetime(data["Date"])

# Sort the DataFrame by date in ascending order
data.sort_values("Date", inplace=True)

# Extract the relevant columns for plotting
price_date = data["Date"]     # X-axis (time series)
price_close = data["Close"]   # Y-axis (Bitcoin closing price)

# Plot the time series
plt.plot_date(price_date, price_close, linestyle="--", color='r')

# Auto-format the x-axis date labels
plt.gcf().autofmt_xdate()

# Add chart title and axis labels
plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

# Adjust layout
plt.tight_layout()

plt.grid(False)

# Save and show the figure
plt.savefig("timeseries.png")
plt.show()
```
