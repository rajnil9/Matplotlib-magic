```python
import csv 
import matplotlib.pyplot as plt
import pandas as pd
```

## Filling areas on lineplot


```python

plt.style.use("ggplot")
# Load the dataset from CSV file
data = pd.read_csv('data.csv')

# Extract required columns into variables
ages = data['Age']              # X-axis (Age of developers)
dev = data['All_Devs']          # Salaries of all developers
py = data['Python']             # Salaries of Python developers
js = data['JavaScript']         # Salaries of JavaScript developers (not plotted here yet)

# Plot line for "All Developers"
plt.plot(ages, dev, color='#444444',
         linestyle='--', label='All Devs')   # Gray dashed line

# Plot line for "Python Developers"
plt.plot(ages, py, label='Python', color="skyblue")  # Skyblue solid line

# Fill area where Python salaries are ABOVE All Dev salaries
plt.fill_between(
    ages, dev, py,                 # Fill between dev and py lines
    where=(py > dev),              # Only fill where Python > All Devs
    alpha=0.3,                     # Transparency level (30% opaque)
    label="Above Avg",             # Legend label
    interpolate=True,              # Ensure smooth shading at crossing points
    color='r'                      # Red shading
)

# Fill area where Python salaries are BELOW All Dev salaries
plt.fill_between(
    ages, dev, py,                 # Fill between dev and py lines
    where=(py <= dev),             # Only fill where Python <= All Devs
    alpha=0.25,                    # Transparency level (25% opaque)
    label="Below Avg",             # Legend label
    interpolate=True,              # Smooth shading at intersections
    color='green'                  # Green shading
)

# Add legend to explain lines and shaded areas
plt.legend()

# Add plot title and axis labels
plt.title('Median Salary Rupees by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary Rupees')

# Adjust layout so labels and titles don’t overlap
plt.tight_layout()
plt.grid(False)
# Save the figure as a PNG file
plt.savefig("areaplot.png")

# Show the final plot
plt.show()
```
