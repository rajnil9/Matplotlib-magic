```python
import matplotlib.pyplot as plt
import numpy as np
```

## Stack Plots

```python

# X-axis values (minutes)
minutes = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# Y-axis values for each player (their contribution over time)
player1 = [8, 6, 5, 5, 4, 2, 1, 1, 0]
player2 = [0, 1, 2, 2, 2, 4, 4, 4, 4]
player3 = [0, 1, 1, 1, 2, 2, 3, 3, 4]

# Labels for legend
labels = ['player1', 'player2', 'player3']

# Custom colors for each player's area
colors = ['#6d904f', '#fc4f30', '#008fd5']

# Create the stacked area plot
plt.stackplot(minutes, player1, player2, player3, labels=labels, colors=colors)

# Add a legend inside the plot (position given by (x, y) coordinates)
plt.legend(loc=(0.07, 0.05))

# Add a title to the chart
plt.title("My Awesome Stack Plot")

# Adjust layout so labels/titles donâ€™t overlap
plt.tight_layout()

# Display the plot
plt.show()

```
