```python
import matplotlib.pyplot as plt
```

## PIE CHARTS

### 1.My initial pie chart 


```python
slices = [120, 80, 30, 40]   # Values for the pie chart (sizes of slices)
l = ["One-twenty", "Eighty", "Thirty", "Forty"]  # Labels for each slice
c = ["red", "blue", "yellow", "orange"]  # Colors for each slice

plt.pie(
    slices,                     # The data (sizes of slices)
    labels=l,                   # Names shown on the chart
    colors=c,                   # Slice colors
    wedgeprops={"edgecolor":"black"}  # Adds black border around each slice
)

plt.title("My 1st pie chart")   # Title of the chart
plt.tight_layout()              # Adjusts layout so nothing overlaps
plt.savefig("pie1.png")
plt.show()                      # Displays the chart
```

### 2.My final pie chart

```python
import matplotlib.pyplot as plt

# Use ggplot style for a nice background and grid
plt.style.use("ggplot")

# Data: number of users for each programming language
slices = [59219, 55466, 47544, 36443, 35917]

# Labels for each slice
l = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

# Colors for the slices
c = ["red", "blue", "yellow", "orange", "green"]

# Explode effect (pull Python slice outward for emphasis)
e = [0, 0, 0, 0.2, 0]

# Create the pie chart
plt.pie(
    slices,                      # Values (sizes of slices)
    labels=l,                    # Labels for each slice
    colors=c,                    # Slice colors
    wedgeprops={"edgecolor":"black"},  # Black border around slices
    shadow=True,                 # Add shadow for 3D effect
    explode=e,                   # Explode (highlight) one slice
    autopct="%1.1f%%",           # Show percentage values inside slices
    startangle=90                # Rotate so chart starts at the top
)

# Add title to the chart
plt.title("Pie chart for Programming Languages")

# Adjust layout so nothing overlaps
plt.tight_layout()

# Save the figure as an image
plt.savefig("pie2.png")

# Display the pie chart
plt.show()

```
