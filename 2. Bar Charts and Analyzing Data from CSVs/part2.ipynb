### 2.Importing csv files

```python
# Import libraries
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt

# Apply ggplot style for visualization
plt.style.use("ggplot")

# Read the CSV file into a DataFrame
data = pd.read_csv("data.csv")

# Extract the 'Responder_id' and 'LanguagesWorkedWith' columns
id = data["Responder_id"]                # Stores respondent IDs
lang = data["LanguagesWorkedWith"]       # Stores languages each person works with (semicolon-separated)

# Create a Counter object to count language frequencies
c = Counter()

# Loop through each response in the 'LanguagesWorkedWith' column
for res in lang:
    # Split by ';' since multiple languages can be listed in one response
    # Example: "Python;JavaScript;SQL"
    c.update(res.split(";"))

# Lists to store languages and their popularity counts
languages = []
pop = []

# Take the 20 most common languages from the Counter
for i in c.most_common(20):
    languages.append(i[0])   # Language name
    pop.append(i[1])         # Number of respondents using it

# Reverse lists so the most popular language appears at the top in the horizontal bar chart
languages.reverse()
pop.reverse()

# Plot a horizontal bar chart
plt.barh(languages, pop)

# Add title and labels
plt.title("Most Popular Languages")
plt.xlabel("Number of People Who Use")
# plt.ylabel("Programming Languages")  # optional, not needed since languages are already on y-axis

# Adjust layout so labels don’t overlap
plt.tight_layout()


plt.savefig("csv file data.png")
# Display the plot
plt.show()
```
